```python
rev_results = roi.back2raw(p4d)
```

```python
rev_results
```

```python
img_dict_sort = p4d.sort_img_by_distance(
    rev_results, roi, 
    num=1  # only keep 3 closest images
)
```

```python
img_dict_sort
```

```python
SAVE_DIR = r"C:\a\crop"

if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

for plt_key in img_dict_sort.keys():
    for img_key in img_dict_sort[plt_key].keys():

        img_coord = img_dict_sort[plt_key][img_key]
        #  
        im_xmin, im_ymin = img_coord.min(axis=0)
        im_xmax, im_ymax = img_coord.max(axis=0)

        print(p4d.photos[img_key].path)

        raw = cv2.imread(str(p4d.photos[img_key].path) , cv2.IMREAD_UNCHANGED)

        if raw is None:
            print('skip')
            continue
#         print(raw)
        raw = raw[int(im_ymin):int(im_ymax), int(im_xmin):int(im_xmax)]

#         plt_id = '1'
        plt_id = plt_key
#         plt_id = plt_key.split('1')[1]
#         if len(plt_key)==1:
        if len(plt_id)==1:
            plt_id = '0' + plt_id
#             plt_id = plt_key.split('_')[0]+'_' + plt_id
        file_name = '0709_'  + plt_id + '_' + img_key + '.png'

        print(file_name)

        SAVE_PATH = os.path.join(SAVE_DIR, file_name)
        print('SAVE_PATH :', SAVE_PATH)

        # image
        cv2.imwrite(SAVE_PATH, raw)
#         break
```
